```python
import pandas as pd

# 1. Charger les données brutes
df_train = pd.read_csv('data/raw/train.csv')

# 2. Inspecter la structure globale
print("--- 5 premières lignes ---")
display(df_train.head()) 

print("\n--- Informations sur les colonnes ---")
df_train.info()
```

```python
# 3. Vérifier la qualité des données (valeurs manquantes et doublons)
print("\n--- Valeurs manquantes par colonne ---")
print(df_train.isnull().sum())

print("\n--- Nombre de lignes en double ---")
print(df_train.duplicated().sum())

# (Optionnel) Pour nettoyer les doublons et les lignes vides :
df_train_clean = df_train.dropna().drop_duplicates()
```

```python
# 4. Analyser la distribution des catégories
# Attention : remplace 'Class Index' par le vrai nom de la colonne qui contient tes labels !
print("\n--- Répartition des articles par catégorie ---")
print(df_train_clean['Class Index'].value_counts())
```

```python
# 1. Créer un dictionnaire pour mapper les chiffres aux vrais noms de catégories (AG News)
label_mapping = {
    1: 'World', 
    2: 'Sports', 
    3: 'Business', 
    4: 'Sci/Tech'
}

# Appliquer le mapping pour créer une nouvelle colonne 'Category'
df_train['Category'] = df_train['Class Index'].map(label_mapping)
```

```python
# 2. Combiner le Titre et la Description en une seule colonne 'Text'
# (Maroine aura ainsi un seul grand texte à nettoyer)
df_train['Text'] = df_train['Title'] + " " + df_train['Description']

# 3. Sélectionner uniquement les colonnes finales dont l'équipe a besoin
df_final = df_train[['Text', 'Category']]

# Vérifier à quoi ressemble notre fichier final
print("--- Aperçu du jeu de données final ---")
display(df_final.head())

# 4. SAUVEGARDER LE FICHIER dans le dossier des données traitées
# C'est CE fichier que tu vas livrer à ton équipe !
df_final.to_csv('data/processed/train_clean.csv', index=False)
```

```python
import pandas as pd

# 1. Charger le fichier de test brut
df_test = pd.read_csv('data/raw/test.csv')

# 2. Définir le mapping des catégories (si tu es dans une nouvelle cellule)
label_mapping = {
    1: 'World', 
    2: 'Sports', 
    3: 'Business', 
    4: 'Sci/Tech'
}

# 3. Appliquer les transformations
# Création de la colonne Category avec les vrais noms
df_test['Category'] = df_test['Class Index'].map(label_mapping)

# Fusion du Titre et de la Description
df_test['Text'] = df_test['Title'] + " " + df_test['Description']

# 4. Conserver uniquement les colonnes utiles
df_test_final = df_test[['Text', 'Category']]

# 5. SAUVEGARDER LE FICHIER de test propre
df_test_final.to_csv('data/processed/test_clean.csv', index=False)

print("Le fichier test_clean.csv a été créé et sauvegardé avec succès ! ")
```
